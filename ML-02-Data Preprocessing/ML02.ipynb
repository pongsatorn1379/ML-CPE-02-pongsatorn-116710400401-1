```python
#พงศธร รอดดี 116710400401-1 Sec2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
#พงศธร รอดดี 116710400401-1 Sec2


#Load Dataset
df = pd.read_csv('StudentScores.csv')
print("LAB1: Dataset Exploration")

#Display Shape
print("\nShape:", df.shape)

#Display Data Types
print("\nData Types:")
print(df.dtypes)

#Display Summary Statistics
print("\nSummary Statistics:")
display(df.describe())

#Display Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

#Display Duplicate Records
print("\nDuplicate Records:", df.duplicated().sum())

#Display Class Distribution
print("\nClass Distribution (Passed):")
print(df['Passed'].value_counts())
```

```python
#พงศธร รอดดี 116710400401-1 Sec2

print("LAB2: Data Visualization")

#Histogram
plt.hist(df['Exam_Score'])
plt.title('Histogram of Exam Score')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()

#Correlation Heatmap
plt.figure(figsize=(6, 4))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#พงศธร รอดดี 116710400401-1 Sec2


print("Part 3: Data Cleaning")

#Duplicate Removal
df = df.drop_duplicates()
print("1. Duplicate Removal: Done")

#Missing Value Handling
df['Study_Hours'] = df['Study_Hours'].fillna(df['Study_Hours'].mean())
print("2. Missing Value Handling: Done")

#Incorrect Data Correction
mean_age = int(df['Age'].mean()) 
df.loc[(df['Age'] < 0) | (df['Age'] > 50), 'Age'] = mean_age
print("3. Incorrect Data Correction: Done")

#Data Type Conversion
df['Age'] = df['Age'].astype(int)
print("4. Data Type Conversion: Age converted to int")

#Compare Mean
mean_val = df['Exam_Score'].mean()
print(f"5. Compare Mean -> Exam_Score Mean: {mean_val:.2f}")

#Compare Median
median_val = df['Exam_Score'].median()
print(f"6. Compare Median -> Exam_Score Median: {median_val:.2f}")
```

```python
#พงศธร รอดดี 116710400401-1 Sec2


print("Part 4: Feature Engineering")

#Label Encoding
le = LabelEncoder()
df['Passed_Encoded'] = le.fit_transform(df['Passed'])
print("1. Label Encoding: Done")

#One-Hot Encoding
df = pd.get_dummies(df, columns=['Gender'], drop_first=True, dtype=int)
print("2. One-Hot Encoding: Done")

# แสดงผลลัพธ์สุดท้าย
display(df.head())
```
